````python
def asci(phrase, r = False):
    """
    Transforme une cha√Æne de caract√®res en enlevant majuscules et accents.
    - Entr√©e : phrase (cha√Æne de caract√®res)
    - Sortie : nouvelle_phrase (cha√Æne de caract√®res)
    """
    if type(phrase) != str:
        raise TypeError('le param√®tre doit √™tre une cha√Æne de caract√®res')
    nouvelle_phrase = ''
    for lettre in phrase:
        if r == False:
            if lettre == '[':
                nouvelle_phrase += ' '
            elif lettre == ']':
                nouvelle_phrase += ' '
            elif lettre == "'":
                nouvelle_phrase += ' '
            elif lettre == ',':
                nouvelle_phrase += ' '
        if r == True:
            if lettre == '{':
                nouvelle_phrase += ' '
            elif lettre == '}':
                nouvelle_phrase += ' '
            elif lettre == "'":
                nouvelle_phrase += ' '
        else:
            nouvelle_phrase += lettre
            
    
    return nouvelle_phrase
````

````python
import json
def save(dico,name):
    tf = open(f"{name}.json", "w")
    json.dump(dico,tf)
    tf.close()
def load(file):
    tf = open(f"{file}.json", "r")
    new_dict = json.load(tf)
    return new_dict
````

````python
def search(phrase, mot):
    o = 0
    for n in range(len(phrase)):
        if phrase[n] == mot[0]:
            for k in range(len(mot)):
                try:
                    if phrase[k+n] == mot[k]:
                        o += 1 
                except IndexError:
                    return False
            if o == len(mot) :
                return True
            else:
                o = 1
    return False
    
          
````

````python
def dur√©e (h,m,s,statut,membre):
    tab = []
    if statut == 'connecte':
        tab.append(membre , h,m,s)
    if statut == 'deconnecte':
        for k in range(len(tab)):
            if membre == tab[k]:
                h = h - tab[k+1]
                m = m - tab[k+2]
                s = s - tab[k+3]
                for n in range(k,k+3):
                    del tab[n]
                    
        return int((int(h),int(m),int(s)))
````
